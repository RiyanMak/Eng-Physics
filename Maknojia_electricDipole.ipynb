```python
import numpy as np
import pylab as pl

# Create data set
w = 0.3
d = 0.1
X = np.linspace(-w, w, 100)
Y = np.linspace(-w, w, 100)
Xm, Ym = np.meshgrid(X, Y)

q1 = int(input("Enter a charge for q1: "))
q2 = int(input("Enter a charge for q2: "))
q1 = q1 * 1e-9
q2 = q2 * 1e-9
k = 8.99 * 10 ** 9

def Ex(x,y):
    f = (k * q1 * x /(x**2 + (y - d/2)**2)**(3/2)) + (k * q2 * x /(x**2 + (y + d/2)**2)**(3/2))
    return f
def Ey(x,y):
    f = (k * q1 * (y - d/2) /(x**2 + (y - d/2)**2)**(3/2)) + (k * q2 * (y + d/2) /(x**2 + (y + d/2)**2)**(3/2))
    return f

def V1(x, y):
    return k * q1 / np.sqrt(x ** 2 + (y - (d / 2)) ** 2)

def V2(x, y):
    return k * q2 / np.sqrt(x ** 2 + (y + (d / 2)) ** 2)

# Create plot
fig = pl.figure(figsize=(10,10))
                          
pl.streamplot(Xm, Ym, Ex(Xm, Ym), Ey(Xm, Ym))
                          
levels = [-200, -90, -50, -35, 0, 35, 50, 90, 200]
cp = pl.contour(Xm, Ym, V1(Xm, Ym) + V2(Xm, Ym), levels, colors='k')
pl.clabel(cp, inline=True, fontsize=10)
pl.show()
```
