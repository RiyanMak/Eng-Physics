```python
import numpy as np
import pylab as pl

# Constants
u0 = 4 * np.pi * 10**-7  # Tm/A
d = 0.1  # meters

def magnetic_field_x(x, y, current1, current2):
    bx1 = -u0 * current1 * (y - d/2) / (2*np.pi*((x**2) + ((y-d/2)**2)))
    bx2 = -u0 * current2 * (y + d/2) / (2*np.pi*((x**2) + ((y+d/2)**2)))
    return bx1 + bx2

def magnetic_field_y(x, y, current1, current2):
    by1 = u0 * current1 * x / (2*np.pi*((x**2) + ((y-d/2)**2)))
    by2 = u0 * current2 * x / (2*np.pi*((x**2) + ((y+d/2)**2)))
    return by1 + by2

# Get user input for currents
current1 = float(input("Enter the current (in Amperes) for the first wire: "))
current2 = float(input("Enter the current (in Amperes) for the second wire: "))

# Set up plot
w = 0.2
x = np.linspace(-w, w, 100)
y = np.linspace(-w, w, 100)
X, Y = np.meshgrid(x, y)

# Calculate magnetic field components using user-defined currents
Bx = magnetic_field_x(X, Y, current1, current2)
By = magnetic_field_y(X, Y, current1, current2)

# Plot with enhancements
pl.figure(figsize=(8, 6))
pl.streamplot(X, Y, Bx, By, density=1, color='k')
pl.plot([0, 0], [-d/2, d/2], 'ro', markersize= 8)  
pl.title("Magnetic Field of Two Parallel Current-Carrying Wires")
pl.xlabel("X (m)")
pl.ylabel("Y (m)")
pl.show()
```
